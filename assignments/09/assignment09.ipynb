### Exercise 1 (Pure Lambda Calculus)

a) *We use the representation of Boolean values in the pure $\lambda$-calculus presented in the lecture, i.e., $True$ is represented as $\lambda xy.x$ and $False$ as $\lambda xy.y$. Using this representation give pure $\lambda$-terms for the following Boolean functions:*

(i) *$xor$, such that e.g. $\overline{xor}\overline{True}\overline{True} \rightarrow^{*}_{\beta} False$ and $\overline{xor}\overline{False}\overline{True} \rightarrow^{*}_{\beta} True$*

Given two arguments $p$ and $q$, we can represent $xor$ as $(p \land \lnot q) \lor (\lnot p \land q)$. We thus first need the pure lambda calculus representations of $or$ and $and$.

---

We can represent $or$ in pure lambda calculus as follows:

\begin{equation*}
    or = \lambda x y.x \; x \; y
\end{equation*}

Intuitively, we have that, if the first argument is $\overline{True}$, the result of the expression after $\beta$-reduction is also $\overline{True}$ since $x \; x \; y$ takes the first argument of $or$ as a result when $x = \overline{True}$; on the other hand, if the first argument is $\overline{False}$, the result after $\beta$-reduction is the second argument of $or$ since $x \; x \; y$ takes the second argument when $x = \overline{False}$.

Similarly, we can represent $and$ in pure lambda calculus as follows:

\begin{equation*}
    and = \lambda x y.x \; y \; x
\end{equation*}

In this case, the result of the expression after $\beta$-reduction is the second argument of $and$ if the first argument is $\overline{True}$ since $x \; y \; x$ takes the second argument when $x = \overline{True}$. If the first argument is $\overline{False}$, the result after $\beta$-reduction is also $\overline{False}$ since $x \; y \; x$ takes the first argument of $and$ as a result when $x = \overline{False}$.

---

Using the representation of $not$ given in the hint, namely $\overline{not} = \lambda a.a \; (\lambda x y.y) \; (\lambda x y.x)$, we can now write $xor$ in pure lambda calculus as

\begin{equation*}
    xor = \lambda pq.(exp_1 \; exp_1 \; exp_2)
\end{equation*}

where

\begin{align*}
    exp_1 &= p \; ((\lambda a.a \; (\lambda x y.y) \; (\lambda x y.x)) \; q) \; p = p \; (\overline{not} \; q) \; p \\
    exp_2 &= ((\lambda a.a \; (\lambda x y.y) \; (\lambda x y.x)) \; p) \; q \; ((\lambda a.a \; (\lambda x y.y) \; (\lambda x y.x)) \; p) = (\overline{not} \; p) \; q \; (\overline{not} \; p)
\end{align*}

(ii) *$implies$, such that e.g. $\overline{implies}\overline{False}\overline{True} \rightarrow^{*}_{\beta} \overline{True}$ and $\overline{implies}\overline{True}\overline{False} \rightarrow^{*}_{\beta} \overline{False}$*

Given two arguments $p$ and $q$, the operator $implies$ is equivalent to $p \; implies \; q \iff \lnot p \; or \; q$, which is the form we will use here. Using the definition of $not$ given in the hint, namely $\overline{not} = \lambda a.a \; (\lambda x y.y) \; (\lambda x y.x)$, and the definition of $or$ from above, $implies$ can be written as follows:

\begin{align*}
    \overline{implies} &= \lambda p q.(((\lambda a.a \; (\lambda x y.y) \; (\lambda x y.x)) \; p) \; ((\lambda a.a \; (\lambda x y.y) \; (\lambda x y.x)) \; p) \; q) \\
    &= \lambda p q.(\overline{not} \ p) \; (\overline{not} \; p) \; q
\end{align*}

(i) *$mult$ and the reduction sequence $\overline{mult} \; \overline{n} \; \overline{m} \rightarrow^{*}_{\beta} \overline{n \cdot m}$*

We can represent multiplication using the following pure lambda calculus formulation:

\begin{equation*}
    \lambda nmfx.n \; ((\lambda pq.p) \; (m \; f \; x)) \; x = \lambda nmfx.n \; (\overline{True} \; (m \; f \; x)) \; x
\end{equation*}

The reduction sequence $\overline{mult} \; \overline{n} \; \overline{m}$ is given below:

\begin{align*}
    &\overline{mult} \; \overline{n} \; \overline{m} = (\lambda nmfx.n \; ((\lambda pq.p) \; (m \; f \; x)) \; x) \; (\lambda gy.g^{n}y) \; (\lambda hz.h^{m}z) \\
    &\rightarrow_{\beta} (\lambda mfx.(\lambda gy.g^{n}y) \; ((\lambda pq.p) \; (m \; f \; x)) \; x) \; (\lambda hz.h^{m}z) \\
    &\rightarrow_{\beta} \lambda fx.(\lambda gy.g^{n}y) \; ((\lambda pq.p) \; ((\lambda hz.h^{m}z) \; f \; x)) \; x \\
    &\rightarrow_{\beta} \lambda fx.(\lambda gy.g^{n}y) \; ((\lambda pq.p) \; ((\lambda z.f^{m}z) \; x)) \; x \\
    &\rightarrow_{\beta} \lambda fx.(\lambda gy.g^{n}y) \; ((\lambda pq.p) \; (f^{m}x)) \; x \\
    &\rightarrow_{\beta} \lambda fx.(\lambda gy.g^{n}y) \; ((\lambda q.f^{m}) \; x) \; x \\
    &\rightarrow_{\beta} \lambda fx.(\lambda gy.g^{n}y) \; f^{m} \; x \\
    &\rightarrow_{\beta} \lambda fx.(\lambda y.f^{nm}y) \; x \\
    &\rightarrow_{\beta} \lambda fx.f^{nm}x = \overline{n \cdot m}
\end{align*}

(ii) *$even$ and the reduction sequence $\overline{even} \; \overline{n} \rightarrow^{*}_{\beta} \overline{not}^n \; \overline{True}$*

We can represent the function $even$ in lambda calculus as follows:

\begin{align*}
    \lambda n.n \; \overline{not} \; \overline{True}
\end{align*}

The reduction sequence $\overline{even} \; \overline{n}$ is given below:

\begin{align*}
    &\overline{even} \; \overline{n} = (\lambda n.n \; \overline{not} \; \overline{True}) \; (\lambda fx.f^{n}x) \\
    &\rightarrow_{\beta} (\lambda fx.f^{n}x) \; (\overline{not} \; \overline{True}) \\
    &\rightarrow_{\beta} (\lambda x.\overline{not}^{n}x) \; \overline{True} \\
    &\rightarrow_{\beta} \overline{not}^{n} \; \overline{True}
\end{align*}

For instance, we have:

\begin{align*}
    &\overline{even} \; \overline{o} = (\lambda n.n \; \overline{not} \; \overline{True}) \; (\lambda fx.x) \\
    &\rightarrow_{\beta} (\lambda fx.x) \; (\overline{not} \; \overline{True}) \\
    &\rightarrow_{\beta} (\lambda x.x) \; \overline{True} \\
    &\rightarrow_{\beta} \overline{True}
\end{align*}

\begin{align*}
    &\overline{even} \; \overline{1} = (\lambda n.n \; \overline{not} \; \overline{True}) \; (\lambda fx.fx) \\

\begin{align*}
